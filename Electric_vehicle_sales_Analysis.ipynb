# Data Preprocessing

```python
# Change column 'Year' Datatype
df['Year']=df['Year'].astype(int)
```

```python
categorical_columns = ['Month_Name', 'State', 'Vehicle_Class','Vehicle_Category', 'Vehicle_Type']
df[categorical_columns] =df[categorical_columns].astype('category')
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['EV_Sales_Quantity']=df['EV_Sales_Quantity'].astype(int)
```

```python
df.info()
```

# Feature Engineering

```python
# Extract components from the Date column
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
# 3-month rolling average of EV sales
df['Sales_Rolling_3'] = df.groupby('State')['EV_Sales_Quantity'].rolling(window=3).mean().reset_index(0, drop=True)
```

df.drop(columns=['Date', 'Month_Name'], inplace=True)

```python
plt.figure(figsize=(6,4))
plt.title('Monthly Analysis of EV Sales in India')
sns.lineplot(x='Month_Name', y='EV_Sales_Quantity', data=df,marker='o', color='r')
plt.xlabel('Month')
plt.ylabel('EV Sales');
```

```python
plt.figure(figsize=(6,7))
plt.title('State-Wise Analysis of EV Sales')
sns.barplot(y='State', x='EV_Sales_Quantity', data=df,hue='State', palette='bright')
plt.xlabel('States')
plt.ylabel('EV Sales');
```

```python
import plotly.graph_objects as go
import pandas as pd

# Prepare year list
years = sorted(df['Year'].unique())
years_with_all = ['All Years'] + years

# Create figure
fig = go.Figure()

# "All Years" data (aggregate across all years)
all_years_data = df.groupby('State')['EV_Sales_Quantity'].sum().reset_index()
fig.add_trace(go.Bar(
    x=all_years_data['EV_Sales_Quantity'],
    y=all_years_data['State'],
    name='All Years',
    orientation='h',
    visible=True
))

# Add a bar for each year
for year in years:
    year_data = df[df['Year'] == year].groupby('State')['EV_Sales_Quantity'].sum().reset_index()
    fig.add_trace(go.Bar(
        x=year_data['EV_Sales_Quantity'],
        y=year_data['State'],
        name=str(year),
        orientation='h',
        visible=False
    ))

# Create dropdown buttons
buttons = []
for i, year in enumerate(years_with_all):
    visibility = [False] * len(years_with_all)
    visibility[i] = True
    buttons.append(dict(label=str(year),
                        method="update",
                        args=[{"visible": visibility},
                              {"title": f"State-Wise EV Sales - {year}"}]))

# Update layout with dropdown
fig.update_layout(
    title="State-Wise EV Sales - All Years",
    updatemenus=[
        dict(
            active=0,
            buttons=buttons,
            direction="down",
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.2,
            yanchor="top"
        )
    ],
    xaxis_title="EV Sales",
    yaxis_title="State",
    height=700
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Unique years sorted
years = sorted(df['Year'].unique().tolist())

# Add 'All Years' at the beginning
years_with_all = ['All Years'] + years

# Create figure
fig = go.Figure()

# "All Years" data
all_years_data = df.groupby('Vehicle_Category')['EV_Sales_Quantity'].sum().reset_index()
fig.add_trace(go.Pie(
    labels=all_years_data['Vehicle_Category'],
    values=all_years_data['EV_Sales_Quantity'],
    name='All Years',
    visible=True  # Only this one is visible initially
))

# Individual year slices
for year in years:
    year_data = df[df['Year'] == year]
    category_sales = year_data.groupby('Vehicle_Category')['EV_Sales_Quantity'].sum().reset_index()
    
    fig.add_trace(go.Pie(
        labels=category_sales['Vehicle_Category'],
        values=category_sales['EV_Sales_Quantity'],
        name=str(year),
        visible=False
    ))

# Dropdown buttons
buttons = []
for i, year in enumerate(years_with_all):
    visibility = [False] * len(years_with_all)
    visibility[i] = True
    buttons.append(dict(label=str(year),
                        method="update",
                        args=[{"visible": visibility},
                              {"title": f"EV Sales by Vehicle Category - {year}"}]))

# Update layout with dropdown
fig.update_layout(
    title="EV Sales by Vehicle Category - All Years",
    updatemenus=[
        dict(
            active=0,
            buttons=buttons,
            direction="down",
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.2,
            yanchor="top"
        )
    ]
)

fig.show()
```
